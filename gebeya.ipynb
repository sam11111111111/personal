```python
from telebot import TeleBot, types
from telebot.types import InlineKeyboardMarkup, InlineKeyboardButton, InputMediaPhoto

bot = TeleBot("7735582459:AAGHKC4ajaAuYf4GKf6ZetTxdPUgxGFtHJw")

# üîπ Define Product Categories
CATEGORIES = {
    "gaminglaptops": [
        {"id": "L002", "name": "HP Omen X", "price": 68010, "image": "L002.jpg"},
        {"id": "L004", "name": "Asus Rog Strix", "price": 122010, "image": "L004.jpg"},
        {"id": "L006", "name": "MSI Stealth", "price": 150000, "image": "L006.jpg"},
        {"id": "L008", "name": "Acer Predator", "price": 98000, "image": "L008.jpg"},
        {"id": "L010", "name": "Razer Blade 15", "price": 170000, "image": "L010.jpg"},
        {"id": "L012", "name": "Alienware M15", "price": 200000, "image": "L012.jpg"},
    ],
    "mobile_phones": [
        {"id": "M001", "name": "Samsung S23 Ultra", "price": 125000, "image": "M001.jpg"},
        {"id": "M002", "name": "iPhone 15 Pro Max", "price": 180000, "image": "M002.jpg"},
        {"id": "M003", "name": "Google Pixel 8", "price": 95000, "image": "M003.jpg"},
        {"id": "M004", "name": "OnePlus 12", "price": 88000, "image": "M004.jpg"},
        {"id": "M005", "name": "Xiaomi 14 Ultra", "price": 110000, "image": "M005.jpg"},
        {"id": "M006", "name": "Vivo X100 Pro", "price": 102000, "image": "M006.jpg"},
    ],
}

# Number of products per page
PAGE_SIZE = 3

@bot.callback_query_handler(func=lambda call: call.data.startswith("category_"))
def show_products(call):
    """Handles category selection & displays the first page of products"""
    category = call.data.split("_")[1]
    chat_id = call.message.chat.id
    send_product_page(chat_id, category, page=0)

def send_product_page(chat_id, category, page):
    """Handles pagination for products within a category"""
    products = CATEGORIES[category]
    start_idx = page * PAGE_SIZE
    end_idx = min(start_idx + PAGE_SIZE, len(products))
    current_page_products = products[start_idx:end_idx]

    # Send product images & descriptions
    media_group = []
    for product in current_page_products:
        caption = f"üì± *{product['name']}*\nüí∞ *Price:* {product['price']} Birr\nüõí Tap 'Order' for details."
        media_group.append(InputMediaPhoto(open(product['image'], "rb"), caption=caption, parse_mode="Markdown"))

    bot.send_media_group(chat_id, media_group)

    # Pagination Buttons
    markup = InlineKeyboardMarkup()

    # Previous Page Button
    if page > 0:
        markup.add(InlineKeyboardButton("‚¨ÖÔ∏è Previous", callback_data=f"page_{category}_{page - 1}"))

    # Next Page Button
    if end_idx < len(products):
        markup.add(InlineKeyboardButton("Next ‚û°Ô∏è", callback_data=f"page_{category}_{page + 1}"))

    # Order Buttons for Each Product
    for product in current_page_products:
        markup.add(InlineKeyboardButton(f"üõí Order {product['name']}", callback_data=f"order_{product['id']}"))

    # Back to Categories Button
    markup.add(InlineKeyboardButton("üîô Back to Categories", callback_data="back_to_categories"))

    bot.send_message(chat_id, f"üìÑ *Page {page + 1} of {len(products) // PAGE_SIZE + 1}*", reply_markup=markup)

@bot.callback_query_handler(func=lambda call: call.data.startswith("page_"))
def paginate(call):
    """Handles page navigation"""
    _, category, page = call.data.split("_")
    send_product_page(call.message.chat.id, category, int(page))

@bot.callback_query_handler(func=lambda call: call.data.startswith("order_"))
def order_product(call):
    """Handles order selection"""
    product_id = call.data.split("_")[1]
    chat_id = call.message.chat.id

    # Find product details
    for category in CATEGORIES.values():
        for product in category:
            if product["id"] == product_id:
                details = (
                    f"üíª *{product['name']} - Full Specifications*\n\n"
                    f"üí∞ **Price:** {product['price']} Birr\n\n"
                    "üöÄ *Ideal for performance & daily tasks!*"
                )
                bot.send_message(chat_id, details)

                # Purchase Button
                markup = InlineKeyboardMarkup()
                markup.add(InlineKeyboardButton("üõí Complete My Purchase", callback_data=f"buy_{product_id}"))
                bot.send_message(chat_id, "Ready to order? üëá", reply_markup=markup)
                return

@bot.callback_query_handler(func=lambda call: call.data == "back_to_categories")
def show_categories(chat_id):
    #chat_id = call.message.chat.id
    """Shows category selection menu"""
    markup = InlineKeyboardMarkup()
    markup.add(InlineKeyboardButton("üéÆ Gaming Laptops", callback_data="category_gaminglaptops"))
    markup.add(InlineKeyboardButton("üì± Mobile Phones", callback_data="category_mobile_phones"))
    bot.send_message(chat_id, "üìå *Select a Product Category:*", reply_markup=markup)

# Command to start
@bot.message_handler(commands=['start'])
def start(message):
    show_categories(message.chat.id)
bot.polling()
```
